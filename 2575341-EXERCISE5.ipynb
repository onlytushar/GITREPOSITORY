# 1Make a bar plot of the months in which movies with "Christmas" in their title tend to be released in the USA.

```python
import matplotlib.pyplot as plt
release_dates = pd.read_csv('release_dates.csv',parse_dates=['date'])

hobbit_movies_usa = release_dates[(release_dates['title'].str.startswith('The Hobbit')) & (release_dates['country'] == 'USA')]
hobbit_movies_usa['month'] = hobbit_movies_usa['date'].dt.month
month_counts = hobbit_movies_usa['month'].value_counts().sort_index()
month_counts.plot(kind='bar', xlabel='Month', ylabel='Number of Releases', title='The Hobbit Movies Released by Month in the USA')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

# 3Make a bar plot of the day of the week on which movies with "Romance" in their title tend to be released in the USA.

```python
romance_movies_usa = release_dates[(release_dates['title'].str.contains('Romance', case=False)) & (release_dates['country'] == 'USA')]
romance_movies_usa['day_of_week'] = romance_movies_usa['date'].dt.dayofweek
day_of_week_counts = romance_movies_usa['day_of_week'].value_counts().sort_index()
day_of_week_counts.plot(kind='bar', xlabel='Day of the Week', ylabel='Number of Releases', title='Romance Movies Released by Day of the Week in the USA')
plt.xticks(range(7), ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.show()
```
